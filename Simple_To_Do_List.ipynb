```python
todo_list = []

def show_menu():
    print("\n=== TO-DO LIST MENU ===")
    print("1. Add a task")
    print("2. Remove a task")
    print("3. View all tasks")
    print("4. Exit")

def add_task():
    task = input("Enter the task to add: ")
    todo_list.append(task)
    print(f"‚úî Task added: {task}")

def remove_task():
    if not todo_list:
        print("‚ùå No tasks to remove.")
        return

    view_tasks()
    try:
        index = int(input("Enter task number to remove: ")) - 1
        if 0 <= index < len(todo_list):
            removed = todo_list.pop(index)
            print(f"‚úî Task removed: {removed}")
        else:
            print("‚ùå Invalid task number.")
    except ValueError:
        print("‚ùå Please enter a valid number.")

def view_tasks():
    if not todo_list:
        print("üì≠ Your to-do list is empty.")
    else:
        print("\n=== Your Tasks ===")
        for i, task in enumerate(todo_list, 1):
            print(f"{i}. {task}")

def todo_app():
    while True:
        show_menu()
        choice = input("Choose an option (1-4): ")

        if choice == "1":
            add_task()
        elif choice == "2":
            remove_task()
        elif choice == "3":
            view_tasks()
        elif choice == "4":
            print("üëã Exiting To-Do List App. Goodbye!")
            break
        else:
            print("‚ùå Invalid choice. Please enter 1-4.")

# Run the app
todo_app()
```
